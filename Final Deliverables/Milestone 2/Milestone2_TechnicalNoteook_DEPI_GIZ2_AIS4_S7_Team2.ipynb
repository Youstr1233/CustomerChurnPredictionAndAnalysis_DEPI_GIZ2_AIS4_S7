Histograms for Numeric Columns
Explanation:
This collection of histograms displays the distribution of values for each numeric column in the dataset. Each histogram shows how frequently each value (or range of values) appears within a column, with the x-axis representing the value ranges and the y-axis showing the frequency (count).
It helps you quickly spot:

If the data is normally distributed, skewed, or uniform.

Where the majority of values fall (central tendency).



```python
data.hist(figsize=(14, 12), bins=20)
plt.suptitle('Histograms for Numeric Columns', fontsize=16)
plt.show()
```

Boxplots for Numeric Columns
Explanation:
These boxplots visualize the spread and summary statistics (like median, quartiles, and outliers) for selected numeric variables: age, estimated_salary, calls_made, sms_sent, and data_used.
Key features of each boxplot:

The middle line shows the median.

The box covers the interquartile range (IQR) (from 25th to 75th percentile).

Whiskers extend to the minimum and maximum values within 1.5 × IQR.

Dots outside the whiskers represent outliers.

```python
for col in ['age', 'estimated_salary', 'calls_made', 'sms_sent', 'data_used']:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

Average Calls Made per Registration Month
Explanation:
This barplot displays the average number of calls made based on the registration month, aggregating all years together.
It’s useful for detecting:

Any monthly trends or seasonality in call behavior.



```python
monthly_calls = data.groupby('registration_month')['calls_made'].mean().reset_index()

# Barplot
plt.figure(figsize=(8, 4))
sns.barplot(data=monthly_calls, x='registration_month', y='calls_made', palette='viridis')
plt.title('Average Calls Made per Registration Month')
plt.ylabel('Average Calls Made')
plt.xlabel('Registration Month')
plt.show()
```

 Average Calls Made per Month in Each Year  , sms sent and data used (Multiple Plots)
Explanation:
This set of barplots visualizes the average number of calls made per registration month within each specific year.
For every year:

It shows the breakdown of monthly call averages.

Helps analyze whether patterns like spikes or drops in call activity are consistent or different across years.

It’s particularly helpful if you have multi-year data to compare month-on-month trends year by year.

```python
# Get unique years from data
years = data['registration_year'].unique()

# Plot for each year
for year in years:
    plt.figure(figsize=(8, 4))

    # Filter data for that year
    year_data = data[data['registration_year'] == year]

    # Barplot for that year
    sns.barplot(data=year_data, x='registration_month', y='calls_made', palette='viridis')

    plt.title(f'Average Calls Made per Month in {year}')
    plt.ylabel('Average Calls Made')
    plt.xlabel('Registration Month')
    plt.ylim(0, data['calls_made'].max() + 10)
    plt.show()
```

Explanation:
This heatmap displays the correlation coefficients between all numeric variables.

Positive values (blue): Variables increase together.

Negative values (red): One variable increases while the other decreases.

Near 0: No linear relationship.

It’s useful for identifying:

Strong relationships (positive or negative) between features.

Variables that might be good predictors for churn.

```python
# Calculate correlation matrix
corr_matrix = data.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()

# Focus on churn correlations
churn_corr = corr_matrix['churn'].sort_values(ascending=False)
print(churn_corr)
```

```python
# Group by Year and Month
year_month_churn = data.groupby(['registration_year', 'registration_month'])['churn'].mean().reset_index()

# Get unique years
years = year_month_churn['registration_year'].unique()

# Plot for each year
for year in years:
    plt.figure(figsize=(8, 4))

    # Filter for that year
    year_data = year_month_churn[year_month_churn['registration_year'] == year]

    # Barplot for months
    sns.barplot(data=year_data, x='registration_month', y='churn', palette='viridis')
    plt.title(f'Average Churn Rate per Month in {year}')
    plt.ylabel('Churn Rate')
    plt.xlabel('Month')
    plt.ylim(0, 1)
    plt.show()
```
