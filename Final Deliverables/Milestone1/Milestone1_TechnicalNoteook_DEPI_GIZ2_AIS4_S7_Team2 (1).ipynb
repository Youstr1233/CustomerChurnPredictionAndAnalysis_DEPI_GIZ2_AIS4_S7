## **4. Handling Outliers**

```python
# Collect numerical columns
df_num = data.select_dtypes(include=[np.number])

# Get Q1, Q3, and IQR
Q1 = df_num.quantile(0.25)
Q3 = df_num.quantile(0.75)
IQR = Q3 - Q1
outlier = (df_num < (Q1 - 1.5 * IQR)) | (df_num > (Q3 + 1.5 * IQR))

# Print outliers in each numerical columns
for i in outlier.columns:
    print(f'Total number of Outliers in column {i} are {outlier[i].sum()}')
```

Despite of __churn column__, the dataset has __not any outliers in any numerical columns__.

## **5. Handling Non-standard Missing Values**

According to __calls_made__, __sms_sent__, and __data_used__ columns , there are unlogical negative values in some entries for one or more of three mentioned columns. These negative values as magnitude are not the same which these negative values can not be replaced by zero.

```python
# Sample of negative calls_made entries in dataset
data[data['calls_made'] < 0]
```

There are __6,713__ entries that have negative calls_made values.
__________________

```python
# Sample of negative sms_sent entries in dataset
data[data['sms_sent'] < 0]
```

There are __7,375__ entries that have negative sms_sent values.
__________________

```python
# Sample of negative data_used entries in dataset
data[data['data_used'] < 0]
```

There are __6,050__ entries that have negative data_used values.
__________________

There are suspected reasons of negative values in the three columns as the following:

  *  __Data Entry Errors__: Human or system errors during data collection or processing might lead to negative values.
  * __Billing Corrections__: If the system initially overcharges a customer and later corrects it, the correction might be logged as a negative usage.
  * __Reversals of Charges__: If a call, SMS, or data usage was mistakenly recorded, then later reversed, it might appear as a negative value.
  * __Refunds or Adjustments__: Some telecom companies issue refunds or adjustments, which might be represented as negative values.
  ___________________________

### **Handling Procedure**

For dealing with calls_made, sms_sent, and data_used of each customer as __magnitude__, and simulating the __non ideal conditions__ that can happen in services transactions, Taking the __absolute value__ of the three columns then adding __a new binary categorical column__ for each column (cals_made, sms_sent, and data_used) that indicates if the value is negative or not will be effective handling procedure in this case.

___________________________________

### **Suspected Columns:**

#### **data_used column**

```python
# Creating negative values of calls_made column
data['calls_made_negative'] = (data['calls_made'] < 0).astype(int)
```

```python
# Taking the absolute value of calls_made
data['calls_made'] = data['calls_made'].abs()
```

#### **sms_sent column**

```python
# Creating negative values of sms_sent column
data['sms_sent_negative'] = (data['sms_sent'] < 0).astype(int)
```

```python
# Taking the absolute value of sms_sent
data['sms_sent'] = data['sms_sent'].abs()
```

#### **data_used column**

```python
# Creating negative values of data_used column
data['data_used_negative'] = (data['data_used'] < 0).astype(int)
```

```python
# Taking the absolute value of data_used
data['data_used'] = data['data_used'].abs()
```
