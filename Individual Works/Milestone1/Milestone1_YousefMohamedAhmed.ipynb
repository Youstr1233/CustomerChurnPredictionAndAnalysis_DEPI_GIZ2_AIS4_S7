## **Handling The Formula of Atribute Data**

Handling formula of attribute data is essential for deeply data analysis and more efficient machine learning model. The way of handling can be __changing the formula__ of data or adding categorical column that __categorize__ a certain column.

### **Suspected Columns:**

#### **age Column**

```python
# Statistical Summary of age column
data['age'].describe()
```

According the statitical summary of age, the age of customers between (18,74) years.
__________________________________

##### **age_decay_range column**

Column that illustrates the __decay range__ of customer.
________________

```python
# Add age_decay_range column
data['age_decay_range'] = data['age'].apply(lambda x: x//10 *10)
```

```python
# Sample of age_decay_range entries
data['age_decay_range']
```

```python
# Categories of age_decay_range column
data['age_decay_range'].value_counts()
```

##### **age_interval column**

Column that illstrate the __age interval__ of the customer.

According to [Facial Age Group Classification Research](https://www.researchgate.net/publication/271253673_Facial_Age_Group_Classification)

![](https://www.researchgate.net/profile/Punit-Raninga/publication/271253673/figure/tbl1/AS:667620638482437@1536184572172/AGE-GROUP-NAME-WITH-ITS-AGE-RANGE.png)

```python
# Add age_interval column
data['age_interval'] = data['age'].apply(lambda x: 'Young Adults' if (x>=17 | x <=30) else('Middle-agged Adults' if (x>=31 | x <=45) else ('Old Adults')) )
```

```python
# Sample of age_interval entries
data['age_interval']
```

```python
# Categories of age_interval column
data['age_interval'].value_counts()
```

##### **region column**

Column that illstrates the __region__ that the customer live in.
_______________________________

According to [Indias By Region](https://learningindia.in/indias-by-region/)

![](https://learningindia.in/wp-content/uploads/2014/04/Slide1.jpg)

```python
# Add region column
data['region'] = data['state'].apply(lambda x:
                                     'North India' if x in ['Rajasthan','Himachal Pradesh','Haryana','Punjab','Uttar Pradesh','Uttarakhand']
                                     else(
                                    'West India' if x in ['Goa','Gujarat','Maharashtra']
                                    else(
                                        'South India' if x in ['Karnataka','Tamil Nadu','Kerala','Andhra Pradesh','Telangana']
                                        else(
                                            'East India' if x in ['Bihar','Jharkhand','Odisha','West Bengal']
                                            else 'Central India' if x in ['Madhya Pradesh','Chhattisgarh']
                                            else 'Northeast India' if x in ['Assam','Manipur','Meghalaya','Mizoram','Nagaland','Sikkim','Tripura','Arunachal Pradesh']
                                            else 'Other'
                                        )
                                    )
                                     )
                                     )
```

```python
# Sample of region entries
data['region']
```

```python
# Categories of region column
data['region'].value_counts()
```
