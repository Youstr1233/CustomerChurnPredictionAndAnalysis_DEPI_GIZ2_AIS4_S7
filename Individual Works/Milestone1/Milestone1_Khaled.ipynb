customer_id

customer_id is the same as the entry position + 1. customer_id will not be used in analysis and building machine learning model. So, customer_id will be dropped.

```python
# Drop customer_id column
data.drop('customer_id', axis=1, inplace=True)
```

pincode

According to "Handling The Formula of Attribute Data" section, pincode data have many contradictions with city and state columns. So, pincode column will be dropped.

```python
# Drop pincode column
data.drop('pincode', axis=1, inplace=True)
```

date of registration column

According to the original formula of date_of_registration, date_of_registration column will be dropped.

```python
# Drop date_of_registration column
data.drop('date_of_registration', axis=1, inplace=True)
```

churn column

churn column is the objective of the whole project. So churn column wil not be dropped.
