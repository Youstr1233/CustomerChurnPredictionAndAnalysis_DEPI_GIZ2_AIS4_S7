```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load dataset
data = pd.read_csv('/content/telecom_churn.csv')

# Convert 'date_of_registration' to datetime format
data['date_of_registration'] = pd.to_datetime(data['date_of_registration'], errors='coerce')
```

check the missing values


```python
data[['estimated_salary', 'date_of_registration']].isna().sum()
```

 Extract year, month, and day as separate features


```python

data['registration_year'] = data['date_of_registration'].dt.year
data['registration_month'] = data['date_of_registration'].dt.month
data['registration_day'] = data['date_of_registration'].dt.day
```

# Drop original date column after extraction


```python
data.drop(columns=['date_of_registration'], inplace=True)
```

# Normalize 'estimated_salary' using MinMaxScaler


```python
scaler = MinMaxScaler()
data['estimated_salary'] = scaler.fit_transform(data[['estimated_salary']])
```
