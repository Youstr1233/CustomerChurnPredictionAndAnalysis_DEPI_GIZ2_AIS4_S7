```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


read_data=pd.read_csv('/content/telecom_churn.csv')



#Before absolute




# Find the most negative values in 'sms sent'
negative_sms = read_data[read_data['sms_sent'] < 0]['sms_sent']
if not negative_sms.empty:
    most_negative_sms = negative_sms.min()
    print(f"The most negative value in 'sms sent': {most_negative_sms}")

    # You can also display a few of the most negative values if there are many
    print("\nFew of the most negative values in 'sms sent':")
    print(negative_sms.sort_values().head())
else:
    print("No negative values found in 'sms sent'.")






# Find the most negative values in 'calls made'
negative_calls = read_data[read_data['calls_made'] < 0]['calls_made']
if not negative_calls.empty:
    most_negative_calls = negative_calls.min()
    print(f"The most negative value in 'calls made': {most_negative_calls}")

    # You can also display a few of the most negative values if there are many
    print("\nFew of the most negative values in 'calls made':")
    print(negative_calls.sort_values().head())
else:
    print("No negative values found in 'calls made'.")














#After absolute




read_data['sms_sent']=read_data['sms_sent'].abs()


read_data['calls_made']=read_data['calls_made'].abs()



# Find the most negative values in 'sms sent'
negative_sms = read_data[read_data['sms_sent'] < 0]['sms_sent']
if not negative_sms.empty:
    most_negative_sms = negative_sms.min()
    print(f"The most negative value in 'sms sent': {most_negative_sms}")

    # You can also display a few of the most negative values if there are many
    print("\nFew of the most negative values in 'sms sent':")
    print(negative_sms.sort_values().head())
else:
    print("No negative values found in 'sms sent'.")






# Find the most negative values in 'calls made'
negative_calls = read_data[read_data['calls_made'] < 0]['calls_made']
if not negative_calls.empty:
    most_negative_calls = negative_calls.min()
    print(f"The most negative value in 'calls made': {most_negative_calls}")

    # You can also display a few of the most negative values if there are many
    print("\nFew of the most negative values in 'calls made':")
    print(negative_calls.sort_values().head())
else:
    print("No negative values found in 'calls made'.")
















```
